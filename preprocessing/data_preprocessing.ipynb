```python
jobplanet_df = pd.read_csv('../data/jobplanet_df.csv',encoding='utf-8')
jobplanet_df
```

```python
import re 
def no_special(text):
    text = text.replace('\n','')
    text = re.sub('[^가-힣a-zA-Z]', ' ', text)
    return text
```

```python
jobplanet_df['요약'] = jobplanet_df['요약'].apply(no_special)
jobplanet_df['장점'] = jobplanet_df['장점'].apply(no_special)
jobplanet_df['단점'] = jobplanet_df['단점'].apply(no_special)
jobplanet_df['경영진에게 바라는 점'] = jobplanet_df['경영진에게 바라는 점'].apply(no_special)
```

```python
jobplanet_df
```

```python
from hanspell import spell_checker

result = spell_checker.check('마춤뻡 적용을위한 테스트입니다')

for i,k in result.as_dict().items():
    print(i+": "+ str(k))
```

```python
need_spellcheck_list = ['요약','장점','단점','경영진에게 바라는 점']

for i in need_spellcheck_list:
    jobplanet_df[i] = jobplanet_df[i].apply(lambda x: spell_checker.check(x).checked)
```

```python
jobplanet_df.to_csv('jobplanet_df_spellcheck.csv',encoding='utf-8',index=False)
```
