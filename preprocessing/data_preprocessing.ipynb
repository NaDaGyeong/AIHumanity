```python
import os
os.getcwd()
```

```python
import pandas as pd 
```

```python
jobplanet_df = os.listdir("../data")
```

```python
df = pd.DataFrame(columns={'회사','날짜','직무','재직여부','별점','요약','장점','단점','경영진에게 바라는 점'})

for i in jobplanet_df:
    temp_df = pd.read_csv("../data/"+i,encoding='utf-8')
    df = pd.concat([df, temp_df])
    
df.info()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.drop(['index'],axis=1,inplace=True)
df.head()
```

```python
df = df[['회사','날짜','직무','재직여부','별점','요약','장점','단점','경영진에게 바라는 점']]
df
```

```python
df.to_csv('jobplanet_df.csv',encoding='utf-8',index=False)
```
